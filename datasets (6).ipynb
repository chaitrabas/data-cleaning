```python
def clean_dates(series, date_format='%d-%m-%Y'):
    """
    Robust date cleaning function that:
    1. Converts to datetime (handling multiple formats)
    2. Fills missing values sensibly
    3. Formats consistently
    """
    # First try converting with dayfirst for European dates
    dates = pd.to_datetime(series, errors='coerce', dayfirst=True)
    
    # For remaining NaT values,
    if dates.isna().any():
        dates = dates.fillna(pd.to_datetime(series, errors='coerce', format='%m/%d/%Y'))
    
    # Fill any still-missing dates with reasonable default
    dates = dates.fillna(pd.to_datetime('today'))
      
    # Convert to desired string format
    return dates.dt.strftime(date_format)

df['Dt_Customer'] = clean_dates(df['Dt_Customer'])
```

```python
def clean_columns(dataframe):
    """
    Standardizes column headers:
    - lowercase
    - underscore separator
    - no special characters
    """
    if dataframe is None:
        raise ValueError("Input DataFrame is None")
    
    new_columns = []
    for col in dataframe.columns:
        # Convert to string if not already
        col = str(col)
        # Convert to lowercase
        col = col.lower()
        # Replace spaces/special characters
        col = col.replace(' ', '_')
        col = ''.join(c if c.isalnum() or c == '_' else '' for c in col)
        # Remove leading/trailing underscores
        col = col.strip('_')
        new_columns.append(col)
    
    dataframe.columns = new_columns
    return dataframe
    
df = clean_columns(df)
print("Cleaned column headers:")
print(df.columns.tolist())
```

```python
from datetime import datetime

def clean_year(year):
    """Convert all years to proper 4-digit format"""
    year = int(year) if not pd.isna(year) else None
    
   # Handle 3-digit years (append 0)
    if 100 <= year < 1000:
        return year * 10
    
    # Keep valid 4-digit years (1900-2025)
    elif 1900 <= year <= datetime.now().year:
        return year
    
    # Return None for invalid years
    else:
        return None

# Apply cleaning
df['year_birth'] = df['year_birth'].apply(clean_year)

# Remove rows with invalid years
df = df.dropna(subset=['year_birth'])

df = df.copy()  # Create new copy
df['year_birth'] = pd.to_numeric(df['year_birth'], errors='coerce')
df = df[df['year_birth'].between(1900, pd.Timestamp.now().year)]
# Verify
print(sorted(df['year_birth'].unique()))
```
